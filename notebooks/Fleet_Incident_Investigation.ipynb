# Fleet Incident Investigation
**Data Analytics & Observability Internship Assignment**

This notebook investigates the operational issues experienced by an Autonomous Mobile Robot (AMR) fleet between 10:00 AM and 11:00 AM. 

Our investigation follows a strict evidence-backed approach based purely on the raw data.

```python
import pandas as pd
import os

# 1. Load Data
telemetry = pd.read_csv('../data/raw/telemetry.csv')
trips = pd.read_csv('../data/raw/trips.csv')

events = []
with open('../data/raw/events.log', 'r') as f:
    for line in f:
        parts = line.strip().split(' ', 3)
        if len(parts) >= 3:
            events.append({
                'timestamp': parts[0],
                'robot_id': parts[1],
                'event_type': parts[2],
                'details': parts[3] if len(parts) > 3 else None
            })
events = pd.DataFrame(events)
print('Data Loaded Successfully.')
```

## Primary Incident: Robot_01 Failure
The most critical issue occurred with Robot_01. By examining the exact rows, we can deduce the root cause.

```python
# Filter evidence for Robot_01
print("--- Event Log ---")
print(events[events['robot_id'] == 'Robot_01'])

print("\n--- Telemetry Snapshot (10:04 to 10:12) ---")
t_r1 = telemetry[telemetry['robot_id'] == 'Robot_01'].copy()
t_r1['time'] = t_r1['timestamp'].astype(str).str[-8:]
print(t_r1[(t_r1['time'] >= '10:04:00') & (t_r1['time'] <= '10:12:00')])

print("\n--- Trip Record ---")
print(trips[trips['robot_id'] == 'Robot_01'])
```
